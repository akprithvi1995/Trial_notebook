# JOBs 

```python
#JOBS - TWO Types of HTML FILES ARE PRESENT
count = 0
files = glob.glob('files/*')
df = pd.DataFrame(columns=['Name','Company','Job Title'])
for file in files:
    
    soup = BeautifulSoup(open(file), "html.parser")
    name = soup.select_one('section.pv-top-card-section .pv-top-card-section__name').get_text().strip()
    job_list = soup.select('.background-section #experience-section ul .pv-position-entity')
    for job in job_list:
        company = job.select_one('.pv-entity__secondary-title').get_text().strip()
        title = job.select_one('.pv-entity__summary-info h3:nth-of-type(1)').get_text().strip()
        df.loc[count] = [name,company,title]
        count+=1
    

soup = BeautifulSoup(open('Bill Stein _ LinkedIn Recruiter.html'),"html.parser")
job_list = soup.select('#profile-experience .position .position-header')
name = soup.select_one('.profile-info h1').get_text()
for job in job_list:
    title = job.select('a')[0].get_text()
    company = job.select('a')[1].get_text()
    df.loc[count] = [name,company,title]
    count+=1
df.to_csv("Job_level.csv",index=False)
```

# Degrees

```python
count = 0
files = glob.glob('files/*')
df1 = pd.DataFrame(columns=['Name','School','Degree'])
for file in files: 
    soup = BeautifulSoup(open(file), "html.parser")
    name = soup.select_one('section.pv-top-card-section .pv-top-card-section__name').get_text().strip()
    deg_list = soup.select('.background-section #education-section .pv-education-entity')
    for degree in deg_list:
        college = degree.select_one('.pv-entity__school-name').get_text().strip()
        deg_name = degree.select_one('.pv-entity__degree-name span:nth-of-type(2)')
        deg_name = deg_name.get_text().strip() if deg_name else "Not Given"
        df1.loc[count]=[name,college,deg_name]
        count+=1


soup = BeautifulSoup(open('Bill Stein _ LinkedIn Recruiter.html'),"html.parser")
deg_list = soup.select('#profile-education .position .position-header')
name = soup.select_one('.profile-info h1').get_text()
for degree in deg_list:
    college = degree.select('a')[0].get_text()
    deg_name = degree.select('h5')[0].get_text()
    df1.loc[count] = [name,college,str(deg_name)]
    count+=1
df1.to_csv("Education_level.csv",index=False)
```
